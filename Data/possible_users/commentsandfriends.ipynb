```python
import networkx as nx
import pandas as pd
import time
import praw
import csv
import prawcore

depression_reddits = ['depressed', 'depression', 'depression_help' , 'depressionrecovery', 'depressionregimens', 'mentalhealth', 'mentalhealthnews', 'mentalhealthuk', 'mentalillness', 'suicide', 'suicidebereavement', 'suicidewatch']

def getCommenters(list_of_comms):
    commentors = list_of_comms.strip('][').split(', ')
    commentors = set(commentors)
    commentors.discard("'None'")
    return commentors

def redditInteractor(list_of_reddits):
    counter = 0
    for subred in list_of_reddits:
        for depred in depression_reddits:
            if subred == depred:
                counter = counter + 1
                break
    return counter/len(list_of_reddits) <= .5



reddit1 = pd.read_csv('./SNAProject/preprocessed_data/depressed.csv')
postersCommenters = reddit1[['author','comments']]
G_Reddit = nx.Graph()



for _, edge in postersCommenters.iterrows():
    G_Reddit.add_node(edge['author'])
    if not type(edge['comments']) == float:
        commentors = getCommenters(edge['comments'])
        for i in commentors:
            i = i[1:-1]
            G_Reddit.add_edge(edge['author'], i)
print(2*G_Reddit.number_of_edges()/G_Reddit.number_of_nodes())
times = reddit1['created_utc']
min_time = times.min()
max_time = times.max()
reddit = praw.Reddit('bot1')
viable_users = list()
for node in list(G_Reddit):
    user = reddit.redditor(node)
    user_subreddits = list()
    try:
        for comment in user.comments.new(limit=None):
            if min_time <= comment.created_utc <= max_time:
                user_subreddits.append(comment.subreddit.display_name)
        for submission in user.submissions.new(limit=None):
            if min_time <= submission.created_utc <= max_time:
                user_subreddits.append(submission.subreddit.display_name)
        if redditInteractor(user_subreddits):
            viable_users.append(node)
    except Exception as e:
        print(e)
print(viable_users)
viable_df = pd.DataFrame(viable_users)
viable_df.to_csv('viable_users.csv')




```
